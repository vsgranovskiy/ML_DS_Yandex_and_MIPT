```python
import pandas as pd
import numpy as np
import math
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

```python
df = pd.read_csv('botswana.tsv', sep = '\t', header = 0)
df.head()
```

```python
print df.columns
```

Загрузите данные и внимательно изучите их. Сколько разных значений принимает признак religion?

```python
print df.religion.value_counts()
```

Во многих признаках есть пропущенные значения. Сколько объектов из 4361 останется, если выбросить все, содержащие пропуски? 

```python
df.index
```

```python
df.dropna()
```

Создайте признак nevermarr, равный единице там, где в agefm пропуски.

```python
new_df = df.copy()
new_df.loc[new_df['agefm'].isnull(),'nevermarr']=1
new_df['nevermarr'].fillna(0,inplace=True)
print new_df.nevermarr.value_counts()
```

```python
new_df.head()
```

Удалите признак evermarr — в сумме с nevermarr он даёт константу, значит, в нашей матрице XXX будет мультиколлинеарность.

```python
new_df.drop(['evermarr'],axis=1, inplace=True)
```

```python
new_df.head()
```

Замените NaN в признаке agefm на cagefm=0c_{agefm}=0cagefm​=0.

```python
new_df.loc[new_df['agefm'].isnull(),'agefm']=0
```

```python
new_df.head()
```

У объектов, где nevermarr = 1, замените NaN в признаке heduc на cheduc1=−1c_{heduc_1}=-1cheduc1​​=−1 (ноль использовать нельзя, так как он уже встречается у некоторых объектов выборки).

```python
new_df.heduc.isna().sum()
```

```python
new_df.loc[(new_df['nevermarr']==1),'heduc']=-1
```

```python
new_df.heduc.isna().sum()
```

```python
new_df.head()
```

Избавимся от оставшихся пропусков.

Для признаков idlnchld, heduc и usemeth проведите операцию, аналогичную предыдущей: создайте индикаторы пропусков по этим признакам (idlnchld_noans, heduc_noans, usemeth_noans), замените пропуски на нехарактерные значения (cidlnchld=−1c_{idlnchld}=-1cidlnchld​=−1, cheduc2=−2c_{heduc_2}=-2cheduc2​​=−2 (значение -1 мы уже использовали), cusemeth=−1c_{usemeth}=-1cusemeth​=−1).

Остались только пропуски в признаках knowmeth, electric, radio, tv и bicycle. Их очень мало, так что удалите объекты, на которых их значения пропущены.

Какого размера теперь наша матрица данных? Умножьте количество строк на количество всех столбцов (включая отклик ceb).

```python
new_df['idlnchld_noans'] = new_df['idlnchld'].apply(lambda x : 1 if (math.isnan(x)) else 0)
new_df['heduc_noans'] = new_df['heduc'].apply(lambda x : 1 if (math.isnan(x)) else 0)
new_df['usemeth_noans'] = new_df['usemeth'].apply(lambda x : 1 if (math.isnan(x)) else 0)
```

```python
values = {'idlnchld': -1, 'heduc': -2, 'usemeth': -1}
```

```python
new_df.fillna(value=values, inplace=True)
```

```python
new_df.head()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.head()
```

```python
print len(new_df.columns)
print len(new_df)
print len(new_df.columns)*len(new_df)
```

```python
new_df.to_csv('pandas.csv', sep='\t')
```

```python
m1 = smf.ols('ceb ~ age + educ + religion + idlnchld + knowmeth + usemeth + agefm +'\
                    'heduc + urban + electric + radio + tv + bicycle + nevermarr +'\
                    'idlnchld_noans + heduc_noans + usemeth_noans', 
             data=new_df)
fitted1 = m1.fit(cov_type='HC1')
print fitted1.summary()
```

```python
print sms.het_breuschpagan(fitted1.resid, fitted1.model.exog)[1]
```

```python
new_df.drop(['religion', 'radio', 'tv'],axis=1, inplace=True)
new_df.head()
```

```python
m2 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + usemeth + agefm +'\
                    'heduc + urban + electric + bicycle + nevermarr +'\
                    'idlnchld_noans + heduc_noans + usemeth_noans', 
             data=new_df)
fitted2 = m2.fit(cov_type='HC1')
print fitted2.summary()
```

```python
print sms.het_breuschpagan(fitted2.resid, fitted2.model.exog)[1]
```

```python
m3 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + usemeth + agefm +'\
                    'heduc + urban + electric + bicycle + nevermarr +'\
                    'idlnchld_noans + heduc_noans + usemeth_noans', 
                    data=new_df)
fitted3 = m3.fit(cov_type='HC1')
print fitted3.summary()
```

```python
print sms.het_breuschpagan(fitted2.resid, fitted2.model.exog)[1]
```

```python
print "F=%f, p=%f, k1=%f" % m1.fit().compare_f_test(m3.fit())
```
